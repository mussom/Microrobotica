```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from Dataframe2AnalisiGrafica import dataframe2analisiGrafica
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('./files/mnist_784_csv.csv')
data
```

```python
listaNomiColonnePixels = list(data.columns)[:-1]
```

```python
cifre = data[listaNomiColonnePixels].values
y = data["class"]
```

```python
fig, ax = plt.subplots(figsize = (3, 3))
ax.imshow(cifre[1].reshape((28, -1)), cmap="binary")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cifre, y, train_size=0.8, random_state=0)
```

```python
model = MLPClassifier(hidden_layer_sizes=(30, 30),
                       random_state=1, max_iter=200)
model.fit(X_train, y_train)

print(f'''
Accuratezza Train: {model.score(X_train, y_train)}
Accuratezza Test: {model.score(X_test, y_test)}
''')
```
