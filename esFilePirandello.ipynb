### funzione file2parole --> testo in lista di parole

```python
def file2parole(filepath, punteggiatura):
    """
    la funzione file2parole trasforma un file di testo nella lista di parole.
    INPUT:
    - filepath: str, path del file di testo
    - punteggiatura: list of str, lista dei caratteri di punteggiatura da eliminare
    
    RETURN
    -lista delle parole
    """
    file = open(filepath, "r", encoding="utf-8")
    righe = file.readlines()
    file.close()
    righeSenzaReturn = [riga[:-1] for riga in righe if len(riga)>1]
    libro = " ".join(righeSenzaReturn).lower()
  
    for carattere in punteggiatura:
        libro = libro.replace(carattere, " ")        

    parole = libro.split(" ")
    parole = [parola for parola in parole if len(parola)>0]
    return parole
    
```

#### creare una funzione che crei un dizionario con la parola(chiave) seguita dalle sue possibili parole [valore(lista)]

## funzione listaParole2Dizionario

```python
def listaParole2Dizionario (lista):
    '''    
    la funzione crea da una lista di parole un dizionario che ha:
        -chiave: ogni parola della lista
        -valore: tutte le parole successive alla chiave che compaiono nel libro

    INPUT:
        -lista:  lista di parole che verrà trasformata in dizionario

    RETURN:
        -dizionario: dizionario con le caratteristiche prima spiegate
    '''
    cont = 0
    diz = {}
    for parola in lista:
        if cont+1 < len(lista):
            if parola not in diz:
                diz[parola] = [lista[cont+1]]
            else:
                diz[parola].append(lista[cont+1])
            cont += 1
    return diz
```

```python
punteggiatura = [",", ";", ".", "'", "*", "(", ")", "!", "?", ":", "«", "»", "-"]
libro = file2parole("./sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", punteggiatura)
dizionarioParole = listaParole2Dizionario(libro)
dizionarioParole
```
